```python
# Cell 4 — inspect the dataset manifest
import json
from pathlib import Path

manifest_path = Path("artifacts/datasets/ds_tinystories_tok_bpe8k_T128/dataset_manifest.json")
manifest = json.loads(manifest_path.read_text())

print(f"Dataset ID    : {manifest['ds_id']}")
print(f"Train tokens  : {manifest['train_tokens']:,}")
print(f"Val tokens    : {manifest['val_tokens']:,}")
print(f"Max seq len   : {manifest['max_seq_len']}")
print(f"Created at    : {manifest['created_at']}")

shard_dir = Path("artifacts/datasets/ds_tinystories_tok_bpe8k_T128")
for shard in sorted(shard_dir.glob("shard_*.bin")):
    size_mb = shard.stat().st_size / 1e6
    print(f"  {shard.name}  {size_mb:.1f} MB")
```
